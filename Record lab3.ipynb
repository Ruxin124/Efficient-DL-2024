```python
import torch
import matplotlib.pyplot as plt
import numpy as np
from models_cifar100.resnet import ResNet18
```

```python
# We load the dictionary
path1 = 'lab3_Resnet18_cifar10_Cosine_16bits.pth'

# loaded_cpt = torch.load(path1)
loaded_cpt = torch.load(path1, map_location=torch.device('cpu'))
print(loaded_cpt.keys())
```

```python
# construct the model
model = ResNet18()

# Load the weights from the saved .pth file
model_weights = loaded_cpt['net']

# Load the state_dict into the model
model.load_state_dict(model_weights)

# Set the model to evaluation mode
model.eval()
```

```python
# Fetch the hyperparam value
hparam_bestvalue = loaded_cpt['hyperparam']
print('hparam_bestvalue:', hparam_bestvalue)
# accuracy
print("best_test_accuracy",loaded_cpt['final_test_accuracy'])
```

```python
# plot the loss on train and test set
plt.figure()
plt.plot(loaded_cpt['epoch_train_losses'], label='train loss')
plt.plot(loaded_cpt['epoch_test_losses'], label='test loss')
plt.legend()
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Loss on train and test set')
plt.show()
```

```python
# plot the accuracy on train and test set
plt.figure()
plt.plot(loaded_cpt['epoch_accuracies_train'], label='train accuracy')
plt.plot(loaded_cpt['epoch_accuracies_test'], label='test accuracy')
plt.legend()
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Accuracy on train and test set')

print('best test accuracy:', loaded_cpt['final_test_accuracy'])
```
