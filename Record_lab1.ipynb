```python
import torch
import matplotlib.pyplot as plt
import numpy as np
```

```python
# We load the dictionary
loaded_cpt = torch.load('Session1_lab2_Resnet18_cifar10.pth', map_location=torch.device('cpu'))
```

```python
# Fetch the hyperparam value
hparam_bestvalue = loaded_cpt['hyperparam']
print('hparam_bestvalue:', hparam_bestvalue)
```

```python
# accuracy
print("",loaded_cpt['final_test_accuracy'])
```

```python
# plot the loss during traing and validation
plt.plot(loaded_cpt['epoch_train_losses'])
plt.plot(loaded_cpt['epoch_test_losses'])
plt.legend(['train', 'test'], loc='upper left')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Loss during training and validation')
```

```python
plt.plot(loaded_cpt['epoch_accuracies_train'])
plt.plot(loaded_cpt['epoch_accuracies_test'])
plt.legend(['train', 'test'], loc='upper left')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Accuracy during training')
```
